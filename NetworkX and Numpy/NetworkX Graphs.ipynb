```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
# setupping the random - seed
seed = 0
random.seed(seed)
np.random.seed(seed)
```

```python
# Adding nodes
g = nx.Graph()
g.add_nodes_from(["A", "B", "C", "D", "E"]) # passing the list of edges.
```

```python
# adding edges
g.add_edges_from(["AC", "BC", "BD", "DE", "CD"])
```

```python
nx.draw(g, node_color = "red", node_size = 750, with_labels = True, font_color = "white", font_size = 20, font_family = "Times New Roman", font_weight = 16, width = 3)
plt.margins(0.5)
plt.show()
```
